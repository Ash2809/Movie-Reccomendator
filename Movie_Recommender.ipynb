```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import ast
from nltk.stem import PorterStemmer
import nltk
```

```python
movies=r"C:\Users\aashutosh kumar\Music\tmdb_5000_movies.csv"
credits=r"C:\Users\aashutosh kumar\Music\tmdb_5000_credits.csv"
```

```python
df_movies=pd.read_csv(movies)
df_credits=pd.read_csv(credits)
```

```python
df_movies.head(3)
```

```python
df_credits.head(3)
```

```python
df_movies=df_movies.merge(df_credits)
df_movies.head(3)
```

```python
df_movies=df_movies.drop(columns=["homepage"])
df_movies.isnull().sum()
```

```python
df_movies.dropna(inplace=True)
print(df_movies.shape)
```

```python
df_movies = df_movies[['movie_id','title','overview','genres','keywords','cast','crew']]
print(df_movies.shape)
df_movies.head(3)
```

```python
# df_movies['crew'][0]
```

```python
def convert(text):
    list=[]
    for i in ast.literal_eval(text):
        list.append(i["name"])
    return list
```

```python
df_movies["genres"]=df_movies["genres"].apply(convert)
print(df_movies.shape)
df_movies.head(3)
```

```python
df_movies["keywords"]=df_movies["keywords"].apply(convert)
print(df_movies.shape)
df_movies.head(3)
```

```python
def convert_cast(text):
    list=[]
    cnt=0
    for i in ast.literal_eval(text):
        cnt+=1
        list.append(i["name"])
        if cnt==3:
            break
    return list
```

```python
df_movies["cast"]=df_movies["cast"].apply(convert_cast)
df_movies.head(3)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df_movies["crew"]=df_movies["crew"].apply(fetch_director)
df_movies.head(3)
```

```python
df_movies['overview'] = df_movies['overview'].apply(lambda x:x.split())
df_movies.head(3)
print(df_movies.shape)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df_movies['cast'] = df_movies['cast'].apply(collapse)
df_movies['crew'] = df_movies['crew'].apply(collapse)
df_movies['genres'] = df_movies['genres'].apply(collapse)
df_movies['keywords'] = df_movies['keywords'].apply(collapse)
```

```python
df_movies["tags"]=df_movies["overview"]+df_movies["genres"]+df_movies["keywords"]+df_movies["cast"]+df_movies["crew"]
print(df_movies.shape)
df_movies.head(3)
```

```python
dataset = df_movies.drop(columns=['overview','genres','keywords','cast','crew'])
dataset['tags'] = dataset['tags'].apply(lambda x: " ".join(x))
print(dataset.shape)
dataset.head()
```

```python
dataset.shape
```

```python
def stemming(text):
    list=[]
    ps=PorterStemmer()
    for i in text.split():
        list.append(ps.stem(i))
    return " ".join(list)
```

```python
dataset["tags"]=dataset["tags"].apply(stemming)
print(dataset.shape)
dataset.head(3)
```

```python
cv=CountVectorizer(stop_words="english",max_features=5000)
vectors=cv.fit_transform(dataset["tags"]).toarray()
```

```python
vectors.shape
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    index = dataset[dataset['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(dataset.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
